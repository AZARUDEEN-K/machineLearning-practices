```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('whitegrid')
```

```python
from sklearn.datasets import make_blobs
data =make_blobs(n_samples=400,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
data[0].shape
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow',marker='o')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,sharey = True,figsize=(10,7))
ax1.set_title('KMeans')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='coolwarm')
ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='coolwarm')
```
